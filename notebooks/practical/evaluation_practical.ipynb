```python
import os
from pathlib import Path
from dotenv import load_dotenv

from retrieval.llm import LLMConfig
from ingestion.image_captioner import VLMConfig
from embedding.embedder import EmbedderConfig
from core.rag import RAG

load_dotenv()
```

```python

rag = RAG(
    storage_dir="./storage",
    postgres_url="postgresql://postgres:postgres@localhost:5432/postgres"
)

rag.set_llm(LLMConfig(
    provider="openai",
    model_name="gpt-4.1-mini",
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL")
))

rag.set_vlm(VLMConfig(
    provider="null",
    model_name="null"
))

rag.create_workspace(
    name="evaluation_practical",
    embedder_config=EmbedderConfig(provider="openai", model_name="text-embedding-3-small")
)
rag.select_workspace("evaluation_practical")
```

```python
RAW_DIR = Path("./raw")

pdf_paths = list(RAW_DIR.glob("*.pdf"))

rag.add_documents([str(p) for p in pdf_paths])

print("Document ingestion complete")
```
