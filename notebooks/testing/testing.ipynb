```python
import os

from core.embedding.embedder import OllamaEmbedder
from core.ingestion.chunker import Chunker
from core.ingestion.image_captioner import ImageCaptioner, OllamaVLM
from core.rag import RAG
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
```

```python
llm = OpenAI(
    api_key="ollama",
    base_url="http://localhost:11434/v1"
)

chunker = Chunker(
    chunk_size=800,
    chunk_overlap=120,
)

image_captioner = ImageCaptioner(
    vlm=OllamaVLM(model_name="moondream"), 
    use_ocr=True
)

rag = RAG(
    storage_dir="./storage",
    llm_client=llm,
    llm_model="llama3.2",  
    chunker=chunker,
    image_captioner=image_captioner,
    postgres_url="postgresql://postgres:postgres@localhost:5432/postgres",
)

embedder_config = OllamaEmbedder(
    model_name="nomic-embed-text", 
    vector_size=768
)
```

```python
rag.create_workspace(name="test", embedder=embedder_config)
rag.select_workspace("test")
```

```python
rag.add_document("./raw/2410.08143v2.pdf")
```
