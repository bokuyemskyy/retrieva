```python
import os


from core.embedding.embedder import OllamaEmbedder
from core.ingestion.chunker import Chunker
from core.ingestion.image_captioner import ImageCaptioner, OllamaVLM
from core.rag import RAG
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
```

```python
llm = OpenAI(api_key=os.getenv("OPENAI_API_KEY"), base_url=os.getenv("OPENAI_BASE_URL"))

embedder = OllamaEmbedder()

chunker = Chunker(
    chunk_size=800,
    chunk_overlap=120,
)

image_captioner = ImageCaptioner(
    vlm=OllamaEmbedder()
)

rag = RAG(
    workspace="test",
    storage_dir="./storage",
    embedder=embedder,
    llm_client=llm,
    chunker=chunker,
    image_captioner=image_captioner,
    llm_model="deepseek-v4-flash",
    postgres_url="postgresql://postgres:postgres@localhost:5432/postgres",
)
```

```python
rag.add_document("./raw/2410.08143v2.pdf")
```
